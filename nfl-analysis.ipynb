```python
qb_games = df[df["name"].isin(qbs)][["name","game_id","epa","yards_gained"]].groupby(["name","game_id"]).sum()
qb_total = df[df["name"].isin(qbs)][["name","game_id","epa","yards_gained"]].groupby("name").sum()
```

```python
plt.figure(figsize=(12,12))
g = sns.scatterplot(x="yards_gained",y="epa",palette="Set1_r",data=qb_total,s=100)
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
g.axhline(0)
g.axvline(qb_total["yards_gained"].mean())


def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.02, point['y'], str(point['val']))

label_point(qb_total.yards_gained, qb_total.epa, qb_total.name, plt.gca()) 
```
